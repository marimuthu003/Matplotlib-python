```python
from matplotlib import pyplot as plt

ages_x=[23,24,25,26,27,28]

#python_developer_salary:
pydev_y=[34555,35999,38000,39766,40766,41700]
plt.plot(ages_x,pydev_y,label='Py devs',color='k',linestyle='--',linewidth=3)

#javascript_developer_salary:
javdev_y=[23888,24877,25888,26788,27777,28777]
plt.plot(ages_x,javdev_y,label='java devs',color='g',linestyle=':',linewidth=3)

#developers salary:
dev_y=[25500,32000,35000,36000,36500,37800]
plt.plot(ages_x,dev_y,label='All devs',color='b',marker='.')

plt.xlabel('Ages of developers')
plt.ylabel('salary of developers (rupees)')
plt.title('Median of salary by age (rupees)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
