# REAL WORLD EXAMPLE USING SCATTER PLOTS


### youtube data view counts and their ratios of the trending videos in 2019 data.

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8')

data=pd.read_csv('2019-05-31-data.csv')
views=data['view_count']
likes=data['likes']
ratio=data['ratio']

plt.scatter(views,likes,c=ratio,cmap='Blues',edgecolors='black',linewidths=0.75,alpha=0.75)

cbar=plt.colorbar()
cbar.set_label('ratios of videos')

#using log to clearly see the data
plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos Data Visulazation')
plt.xlabel('viewcounts')
plt.ylabel('likes')

plt.tight_layout()

plt.show()
```
