```python
from matplotlib import pyplot as plt
data=pd.read_csv('data2.csv')
ages=data['Age']
all_dev=data['All_Devs']
py_dev=data['Python']
js_dev=data['JavaScript']
plt.plot(ages,all_dev,linestyle='--',color='k',label='all_dev')
plt.plot(ages,py_dev,label='python')
#plt.fill_between(ages,py_dev)
overallmedian=57287
plt.fill_between(ages,py_dev,overallmedian,where=(py_dev>overallmedian),interpolate=True,alpha=0.25)
plt.fill_between(ages,py_dev,overallmedian,where=(py_dev<=overallmedian),interpolate=True,alpha=0.25)
#plt.fill_between(ages,py_dev,overallmedian,where=(py_dev>all_dev),interpolate=True,alpha=0.25)
#plt.fill_between(ages,py_dev,overallmedian,where=(py_dev<=all_dev),interpolate=True,alpha=0.25)

plt.title('developers salaries')
plt.xlabel('ages')
plt.ylabel('salaries of developers')
plt.tight_layout()
plt.legend()
plt.show()
```
