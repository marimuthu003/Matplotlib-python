```python
from matplotlib import pyplot as plt
minutes=[1,2,3,4,5,6,7,8,9]

player1=[1,2,2,3,3,4,4,5,6]
player2=[3,4,2,3,5,2,3,4,7]
player3=[1,7,4,8,3,2,4,9,5]

labels=['player1','player2','player3']
colors=['#d4d135','#46bf3d','#3db7bf']
plt.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)
plt.legend(loc='upper left')#legend on upper left
#plt.legend(loc='lower left')#legeng on lower left
#plt.legend(loc=((0.1,0.2)))#legend according to our value by using the tuple
plt.tight_layout()
plt.title('Sample Stack Plot')
plt.show()
```
