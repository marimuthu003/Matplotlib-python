```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
slices=[80,20]
labels=['diet','excercise']
color=['#c4314a','#f5b3be']
explode=[0,0.1]
plt.pie(slices,labels=labels,colors=color,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True,startangle=90,autopct='%1.1f%%')
plt.title('Sample Pie Chart Example')
plt.show()
```
