```python
import numpy as np
from matplotlib import pyplot as plt


ages_x=[23,24,25,26,27,28]
x_indexes=np.arange(len(ages_x))
width=0.25

#python_developer_salary:
pydev_y=[34555,35999,38000,39766,40766,41700]
plt.bar(x_indexes-width,pydev_y,width=width,label='Py devs',color='g')

#javascript_developer_salary:
javdev_y=[23888,24877,25888,26788,27777,28777]
plt.bar(x_indexes,javdev_y,width=width,label='java devs',color='k')

#developers salary:
dev_y=[25500,32000,35000,36000,36500,37800]
plt.bar(x_indexes+width,dev_y,width=width,label='All devs',color='b')

plt.xlabel('Ages of developers')
plt.ylabel('salary of developers (rupees)')
plt.title('Median of salary by age (rupees)')
plt.legend()
#plt.grid(True)
plt.tight_layout()
plt.xticks(ticks=x_indexes,labels=ages_x)
plt.show()
```

# MOST COMMON LANGUAGES USED BY DEVELOPERS:

# PLOTTING THE BAR CHART:

```python
import numpy as np
import csv
from collections import Counter
with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()

    for row in csv_reader:
      language_counter.update(row['LanguageHaveWorkedWith'].split(';'))
languages=[]
popularity=[]     
for item in language_counter.most_common(15):
   languages.append(item[0])
   popularity.append(item[1])
#print(languages)
#print(popularity)
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)#horizontal bar
#plt.ylabel('most popular languages')
plt.xlabel('number of people who use')
plt.title('Programming languages')
plt.tight_layout()
plt.show()
```

# PLOTTING USING PANDAS:


```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter

data=pd.read_csv('data.csv')
index=data['ResponseId']

lang_response=data['LanguageHaveWorkedWith']
language_counter=Counter()

for response in lang_response:
   if isinstance(response, str):
        language_counter.update(response.split(';'))


languages=[]
popularity=[]
for item in language_counter.most_common(15):
   languages.append(item[0])
   popularity.append(item[1])
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)#horizontal bar
#plt.ylabel('most popular languages')
plt.xlabel('number of people who use')
plt.title('Programming languages')
plt.tight_layout()
plt.show()
```
