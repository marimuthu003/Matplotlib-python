```python
import csv
import time
import random

x_value=0
total_1=1000
total_2=1000

fieldnames=['x_value','total_1','total_2']
with open('data4.csv','w') as csv_file:
    csv_writer=csv.DictWriter(csv_file,fieldnames=fieldnames)
    csv_writer.writeheader()

while True:
    with open('data4.csv','a') as csv_file:
        csv_writer=csv.DictWriter(csv_file,fieldnames=fieldnames)

        info = {
            "x_value": x_value,
            "total_1": total_1,
            "total_2": total_2
        }

        csv_writer.writerow(info)
        print(x_value, total_1, total_2)

        x_value+=1
        total_1=total_1+random.randint(-6,8)
        total_2=total_2+random.randint(-5,6)
    time.sleep(1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

fig,ax1=plt.subplots()
def animate(i):
   data=pd.read_csv('data4.csv')
   x_vals=data['x_value']
   y1=data['total_1']
   y2=data['total_2']

   ax1.clear()

   
   ax1.plot(x_vals, y1, label='Channel 1')
   ax1.plot(x_vals, y2, label='Channel 2')

   ax1.legend(loc='upper left')


ani=FuncAnimation(fig,animate,interval=10000)
# Display the animation in Jupyter Notebook
from IPython.display import HTML
HTML(ani.to_jshtml())

# Adjust subplot parameters to make the plot fit into the figure area
fig.tight_layout()
```
